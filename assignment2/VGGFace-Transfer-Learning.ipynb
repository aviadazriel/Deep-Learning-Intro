```python
%matplotlib inline
import src.lfw_dataset as lfw
from src.lfw_dataset import LFWDataLoader
from src.siamese import Siamese

# load the paths of the data sets
same_train_paths, diff_train_paths, same_val_paths, diff_val_paths, same_test_paths, diff_test_paths = lfw.load_data()

def try_train(siamese_net, epochs=400, epoch_shuffle=True, verbose=0):
    """
    Try to train the siamese_net model. If success, return train history. Otherwise, return None.
    """
    try:
        print("Training...")
        history = siamese_net.train(same_train_paths, diff_train_paths, same_val_paths, diff_val_paths, epochs=epochs, epoch_shuffle=epoch_shuffle, verbose=verbose, use_worst_pairs=False)
        print("Training End")
        return history
    except Exception as e:
        print(f"Got exception while training: {e}")
        return None
```

VGG16

```python
s_net = Siamese()
s_net.build('vggface', use_second_dense_layer=True, dense_size_1=2048, dense_size_2=1024, pre_trained_model='vgg16' )
print("Model:")
s_net.model.summary()

history = try_train(s_net, verbose=1)
if history:
    s_net.evaluate(history, same_test_paths, diff_test_paths)
```

RESNET50 

```python
s_net = Siamese()
s_net.build('vggface', dense_size_1=2048, learning_rate=1e-3, dropout_prob=0.2, pre_trained_model='resnet50' )
print("Model:")
s_net.model.summary()
    
history = try_train(s_net, verbose=1)
if history:
    s_net.evaluate(history, same_test_paths, diff_test_paths)
```
