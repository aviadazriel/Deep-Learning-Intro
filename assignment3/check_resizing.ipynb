```python
import matplotlib.pyplot as plt
%matplotlib inline
import pretty_midi
import cv2
```

```python
def extract_midi_piano_roll(midi_path, resize_time=128, fs=10):
    midi_obj = pretty_midi.PrettyMIDI(midi_path)
    results = midi_obj.get_piano_roll(fs=fs)

    resized = cv2.resize(results, (resize_time, results.shape[0]))

    return results, resized
```

```python
original, resized = extract_midi_piano_roll('Data/midi_files/Cat_Stevens_-_Wild_World.mid')
```

```python
plt.figure(figsize=(50,10))
plt.imshow(original)
```

```python
plt.figure(figsize=(50,10))
plt.imshow(resized)
```
